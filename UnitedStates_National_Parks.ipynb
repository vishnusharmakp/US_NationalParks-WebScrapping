#**Web Scrapping Program**


```python
import requests
import re
import pandas as pd
from bs4 import BeautifulSoup
```

*Import all the libraries required*

```python
page = requests.get("https://www.nps.gov/index.htm")
soup = BeautifulSoup(page.content, 'html.parser')
# print(soup.prettify())
```

This goes through the webpage and can read the html code.
We can print by uncommenting the print statement.

```python
state_name = []
for ultag in soup.find_all('ul', {'class': 'dropdown-menu SearchBar-keywordSearch'}):
  for litag in ultag.find_all('li'):
    names = (litag.text)
    for links in litag.find_all('a'):
      if 'href' in links.attrs:
        state_list = (str("https://www.nps.gov"+links.attrs['href']))
        print(f'{names} : {state_list}')
        state_name.append(state_list)
```

This loop above gets all the links from the drop down box which is the *State List*

```python
# print(len(state_name))
# state_name
```

Can Print and verify the same

```python
alink = []
aName = []
aCat = [] 
pDisc = []     
for link in state_name:
  page = requests.get(link)
  soup = BeautifulSoup(page.content, 'html.parser')
  # Find all the divs we need in one go.
  divs = soup.find_all('div', class_='list_left')
  for div in divs:
    # find all the enclosing a tags.
    pcategory = div.find('h2')
    aCat.append(pcategory.get_text())

    pdescription = div.find('p')
    pDisc.append(pdescription.get_text().strip())

    anchors = div.find_all('a')
    for anchor in anchors:
      alink.append("https://www.nps.gov"+anchor.get('href')+"index.htm")
      aName.append(anchor.text.strip())
```

Create an empty list or array in order to store all the necessary information.

```python
print(alink)
print(aName)
print(aCat)
print(pDisc)
# print(len(aName))
# print(len(alink))
# print(len(aCat))
# print(len(pDisc))
```

Print to check if information is right. Can print the length of the list as well to make sure we are having all the enteries.

```python
pAddress = []
pLocality = [] 
pRegion = []
pPostal = []
pTelephone = []

for link1 in alink:
  parkPage = requests.get(link1)
  parkSoup = BeautifulSoup(parkPage.content, 'html.parser')

  # Find all the divs we need in one go.
  pSpans = parkSoup.find_all('div', class_='vcard')

  for span in pSpans:
    pAdr = span.find('span', itemprop="streetAddress")
    if (pAdr is not None):
      pAddress.append(pAdr.get_text().strip('\n'))
    else:
      pAddress.append("")
    
    pLoc = span.find('span', itemprop="addressLocality")
    if (pLoc is not None):
      pLocality.append(pLoc.get_text().strip())
    else:
      pLocality.append("")
    
    pReg = span.find('span', itemprop="addressRegion")
    if (pReg is not None):
      pRegion.append(pReg.get_text().strip())
    else:
      pRegion.append("")
    
    pCode = span.find('span', itemprop="postalCode")
    if (pCode is not None):
      pPostal.append(pCode.get_text().strip())
    else:
      pPostal.append("")
    
    pNum = span.find('span', itemprop="telephone")
    if (pNum is not None):
      pTelephone.append(pNum.get_text().strip())
    else:
      pTelephone.append("")

```

This is to fetch all the necessary information of the address and contact number

```python
aLine1 = []
aLine2 = []
aLine3 = []

for aline in pAddress:
  final = aline.split('\n')
  if(len(final)==1):
    aLine1.append(final[0])
    aLine2.append("")
    aLine3.append("")
  elif(len(final)==2):
    aLine1.append(final[0])
    aLine2.append(final[1])
    aLine3.append("")
  elif(len(final)==3):
    aLine1.append(final[0])
    aLine2.append(final[1])
    aLine3.append(final[2])
  else:
    pass
```

This splits the address into addressline 1, 2, and 3

```python
colNames = {'Park Name':aName, 'Park Category':aCat, 'Description':pDisc, 'Address Line 1':aLine1, 'Address Line 2':aLine2, 'Address Line 3':aLine3, 'City':pLocality, 'State':pRegion, 'Zip Code':pPostal, 'Phone Number':pTelephone}
parkFile = pd.DataFrame(colNames)

parkFile.to_csv('National-Parks-List.csv', index=False)
```
